```python
df['buying'],_=pd.factorize(df['buying'])#we have to use two variable , (,_) act as one variable,which not used in coding..
df['maint'],_=pd.factorize(df['maint'])
df['doors'],_=pd.factorize(df['doors'])
df['persons'],_=pd.factorize(df['persons'])
df['lug_boot'],_=pd.factorize(df['lug_boot'])
df['safety'],_=pd.factorize(df['safety'])
df.head(10)
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)#random_state =0 avoid takes random values,
#it takes train and test data continously and not randomlly
```

```python
x.columns
```

```python
from sklearn import tree,metrics#used find quality of split =>gini
dtree = tree.DecisionTreeClassifier(criterion='gini')
dtree.fit(x_train,y_train)
```

```python
y_train_pred=dtree.predict(x_train)
```

```python
accuracy=metrics.accuracy_score(y_train,y_train_pred)
print('Train accuracy:{:.2f}'.format(accuracy))
```

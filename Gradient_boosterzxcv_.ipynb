```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
#load data
test = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/test.csv")
train = pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Titanic/train.csv")
```

```python
#in colab we can use commas for running multiple commands
train.info(),test.info()
```

```python
train
```

```python
#passengerid are replace with index...
train.set_index("PassengerId",inplace=True)
test.set_index("PassengerId",inplace=True)
train.head(10)
```

```python
y_train = train["Survived"]#this is used for predictions
```

```python
train.drop(labels="Survived",axis=1,inplace=True)#in train date we we,take all data except survived(output)
train
```

```python
train.shape
```

```python
train_test = train.append(test)
```

```python
train_test.shape
```

```python
#delete columns that are nit used as features for training and prediction
columns_to_drop = ["Name","Age","SibSp","Ticket","Cabin","Parch","Embarked"]#our wish
train_test.drop(labels=columns_to_drop,axis=1,inplace=True)
```

```python
train_test
```

```python
train_test.isnull().sum()
```

```python
train_test_dummies = pd.get_dummies(train_test,columns=["Sex"])
train_test_dummies
```

```python
train_test_dummies.shape
```

```python
train_test_dummies.fillna(value=0.0,inplace=True)
train_test_dummies.isnull().sum()
```

```python
X_train=train_test_dummies.values[0:891]
X_test=train_test_dummies.values[891:]
```

```python
#we use min-max scalce ->fare contains large values
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)#we already done fitting ,so do need again
```

```python
# split training feature and target sets into training and validation subsets
from sklearn.model_selection import train_test_split

X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(X_train_scale, y_train, random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
learning_rates = [0.05, 0.1, 0.25, 0.5, 0.75, 1]
for learning_rate in learning_rates:
    gb = GradientBoostingClassifier(n_estimators=20, learning_rate = learning_rate, max_features=2, max_depth = 2, random_state = 0)
    gb.fit(X_train_sub, y_train_sub)
    print("Learning rate: ", learning_rate)
    print("Accuracy score (training): {0:.3f}".format(gb.score(X_train_sub, y_train_sub)))
    print("Accuracy score (validation): {0:.3f}".format(gb.score(X_test_sub, y_test_sub)))
    print()
```
